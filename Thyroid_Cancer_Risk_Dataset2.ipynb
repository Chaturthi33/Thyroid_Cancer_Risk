# HeatMap

```python
import seaborn as sns
data = {'Age': [66, 29, 86, 75, 35],
        'TSH_Level': [9.37, 1.83, 6.26, 4.10, 9.10],
        'T3_Level': [1.67, 1.73, 2.59, 2.62, 2.11],
        'T4_Level': [6.16, 10.54, 10.57, 11.04, 10.71],
        'Nodule_Size': [1.08, 4.05, 4.61, 2.46, 2.11]
}
df = pd.DataFrame(data)

#  correlation matrix
corr = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Thyroid Dataset Features')
plt.show()
```

# Histogram Plot

```python
data = { 'Age': [66, 29, 86, 75, 35, 58, 89, 72, 85, 46],
         'TSH_Level': [9.37, 1.83, 6.26, 4.10, 9.10, 2.00, 9.77, 7.72, 5.62, 5.60],
         'T3_Level': [1.67, 1.73, 2.59, 2.62, 2.11, 0.64, 3.25, 2.44, 2.53, 2.73],
         'Nodule_Size': [1.08, 4.05, 4.61, 2.46, 2.11, 1.48, 4.46, 2.36, 1.54, 2.53]
}
df = pd.DataFrame(data)
fig, axs = plt.subplots(2,2, figsize=(10, 10))  #Creates a 2x2 grid of subplots â€” a total of 4 individual plots (axes) arranged in 2 rows and 2 columns.
                                                #10,10=10 inches wide and 10 inches tall
axs[0, 0].hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Age Distribution')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Frequency')

axs[0, 1].hist(df['TSH_Level'], bins=10, color='lightgreen', edgecolor='black')
axs[0, 1].set_title('TSH Level Distribution')
axs[0, 1].set_xlabel('TSH Level')
axs[0, 1].set_ylabel('Frequency')

axs[1, 0].hist(df['T3_Level'], bins=10, color='salmon', edgecolor='black')
axs[1, 0].set_title('T3 Level Distribution')
axs[1, 0].set_xlabel('T3 Level')
axs[1, 0].set_ylabel('Frequency')

axs[1, 1].hist(df['Nodule_Size'], bins=10, color='plum', edgecolor='black')
axs[1, 1].set_title('Nodule Size Distribution')
axs[1, 1].set_xlabel('Nodule Size')
axs[1, 1].set_ylabel('Frequency')


plt.tight_layout()
plt.show()
```

# Pair Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {'Age': [66, 29, 86, 75, 35],
        'TSH_Level': [9.37, 1.83, 6.26, 4.10, 9.10],
        'T3_Level': [1.67, 1.73, 2.59, 2.62, 2.11],
        'T4_Level': [6.16, 10.54, 10.57, 11.04, 10.71],
        'Nodule_Size': [1.08, 4.05, 4.61, 2.46, 2.11],
        'Thyroid_Cancer_Risk': ['Low', 'Low', 'Low', 'Medium', 'High']}
df = pd.DataFrame(data)
# Optional: Convert 'Thyroid_Cancer_Risk' to category if needed
df['Thyroid_Cancer_Risk'] = pd.Categorical(df['Thyroid_Cancer_Risk'], categories=['Low', 'Medium', 'High'], ordered=True)
sns.pairplot(df, vars=['Age', 'TSH_Level', 'T3_Level', 'T4_Level', 'Nodule_Size'], hue='Thyroid_Cancer_Risk')
plt.suptitle("Pair Plot of Thyroid Dataset Features Colored by Cancer Risk", y=1.02)
plt.show()
```
